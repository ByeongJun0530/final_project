```python
import pandas as pd
```

```python
df1 = pd.read_csv('train_1.csv')
df2 = pd.read_csv('train_2.csv')
```

```python
# train concat
start_idx = len(df1)  # 598
df2['PRODUCT_ID'] = [f'TRAIN_{start_idx + i:03d}' for i in range(len(df2))]

# 합치기
df_concat = pd.concat([df1, df2], axis=0, ignore_index=True)

print('shape:', df_concat.shape)

df_concat.drop(columns=['label_source'], inplace=True)

df_concat.to_csv('train.csv', index=False)
```
